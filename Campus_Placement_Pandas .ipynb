```python
# Find the distribution
# one column
print(df['salary'].skew())  """For numeric columns only output - 0 --> normal distribution, otherwise it is considered as Skew distribution
                              if the distribution is skew use median to fill null values """
```

```python
median_salary = df["salary"].median()

print(median_salary)
```

```python
df["salary"] = df["salary"].fillna(median_salary)
```

```python
columns = df.select_dtypes(include=['int64', 'float64']).columns
columns
```

```python
# find Outlier --- only for numerical columns 
columns=['sl_no', 'ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'salary']
```

```python
# Find Outliers
for col in columns:
    q1=df[col].quantile(0.25)  # Quantile - first 25 % data mean
    q3=df[col].quantile(0.75) # Quantile - first 75% data mean
    iqr=q3-q1 # inner quantile
    lower_bound=q1 - 1.5 * iqr
    upper_bound=q3 + 1.5 * iqr
    Outliers=df[(df[col]>upper_bound) | (df[col]<lower_bound)]
    print(len(Outliers))
    
```

```python
# Handle Outliers with clip function
for col in columns:

    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)

    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    df[col] = df[col].clip(lower=lower_bound,
                           upper=upper_bound)
```

```python
# Recheck the ouliers
for col in columns:
    q1=df[col].quantile(0.25)  # Quantile - first 25 % data mean
    q3=df[col].quantile(0.75) # Quantile - first 75% data mean
    iqr=q3-q1 # inner quantile
    lower_bound=q1 - 1.5 * iqr
    upper_bound=q3 + 1.5 * iqr
    Outliers=df[(df[col]>upper_bound) | (df[col]<lower_bound)]
    print(len(Outliers))
    
```
